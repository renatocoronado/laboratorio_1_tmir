```python
# cargar librerias

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Problema 1

```python
# leer imagen y convertir a formato correcto de color

img1 = cv2.imread("img_1.jpg")  

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
# vista previa imagen original

plt.imshow(img1)
plt.show()
```

```python
# setear dimensiones

alto = img1.shape[0]
ancho = img1.shape[1]
```

```python
# 

nueva_img1 = np.zeros((alto, ancho, 3))
```

```python
# crear funcion

def funcion_1(imagen, color):

    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = imagen[i, j]

            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]

            if color == 1:
                nueva_img1[i, j] = [blue, 0, 0]
            elif color == 2:
                nueva_img1[i, j] = [0, green, 0]
            elif color == 3:
                nueva_img1[i, j] = [0, 0, red]
            elif color == 10:
                nueva_img1[i, j] = [0, green, red]
            elif color == 20:
                nueva_img1[i, j] = [blue, green, 0]
            elif color == 30:
                nueva_img1[i, j] = [blue, 0, red]
            
    return nueva_img1
```

```python
# llamar funcion

funcion_1(img1, 3)
```

```python
# escribir resultado

cv2.imwrite("nueva_img1.jpg", nueva_img1)
```

Problema 3

```python
# leer imagen y convertir a formato correcto de color

img3 = cv2.imread("img_problema_3.jpg")  

img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
```

```python
# crear funcion

def funcion_3(imagen):

    b, g, r = cv2.split(imagen)

    gris_b = cv2.cvtColor(b, cv2.COLOR_GRAY2RGB)
    gris_g = cv2.cvtColor(g, cv2.COLOR_GRAY2RGB)
    gris_r = cv2.cvtColor(r, cv2.COLOR_GRAY2RGB)

    return gris_b, gris_g, gris_r
```

```python
# Crear las tres imágenes en escala de grises por canales

gris_b, gris_g, gris_r = funcion_3(img3)
                    
```

```python
# escribir imagenes

cv2.imwrite("nueva_img3_blue.jpg", gris_b)
cv2.imwrite("nueva_img3_red.jpg", gris_r)
cv2.imwrite("nueva_img3_green.jpg", gris_g)
```

Problema 4

```python
# leer imagen y convertir a formato correcto de color

img4 = cv2.imread("img_1.jpg")  

img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB) 
```

```python
# vista previa imagen original

plt.imshow(img4)
plt.show()
```

```python
# dimensiones

alto = img4.shape[0]
ancho = img4.shape[1]
```

```python
# crear imagen en escala de grises

img4_g = np.mean(img4, axis=-1).astype(np.uint8)
```

```python
# crear funcion

def funcion_4(imagen, imagen_gris):

    colors = ('b','g','r')
    
    for i,color in enumerate(colors):
        hist = cv2.calcHist([imagen],[i],None,[256],[0,256])
        plt.axvline(np.mean(hist), color=color, linestyle='dashed', linewidth=1)
        plt.plot(hist,color = color)

    hist_g = cv2.calcHist([imagen_gris],[0],None,[256],[0,256])
    plt.axvline(np.mean(hist_g), color='gray', linestyle='dashed', linewidth=1)
    plt.plot(hist_g,color = 'gray')


    plt.title('Histograma RGB y Gris')
    plt.show()
```

```python
# llamar funcion

funcion_4(img4, img4_g)
```
